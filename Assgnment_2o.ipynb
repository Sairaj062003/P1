```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dat_set['sex'] = le.fit_transform(dat_set['sex'])
dat_set['smoker'] = le.fit_transform(dat_set['smoker'])
```

```python
dat_set
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False)  


encoded_regions = encoder.fit_transform(dat_set[['region']])


encoded_df = pd.DataFrame(encoded_regions, columns=encoder.get_feature_names_out(['region']))


dat_set = pd.concat([dat_set, encoded_df], axis=1)


dat_set = dat_set.drop(columns=['region'])

```

```python
dat_set
```

```python
from sklearn.model_selection import train_test_split 
X = dat_set[['age']]
y = dat_set[['charges']].values  
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=40)
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
y_predict = lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error

print("MAE ",mean_absolute_error(y_test,y_predict))
```

```python
from sklearn.metrics import mean_squared_error
print("MSE ",mean_squared_error(y_test,y_predict))
```

```python
from sklearn.metrics import r2_score
print("R2 Score ",r2_score(y_test,y_predict))
```

```python
features= ['sex','age','bmi','children','smoker','region_northeast','region_northwest','region_southeast','region_southwest']
target = 'charges'
X = dat_set[features]
y = dat_set[target]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, train_size=0.75, random_state=42)


lr = LinearRegression()


lr.fit(X_train, y_train)


y_predict = lr.predict(X_test)


r2 = r2_score(y_test, y_predict)
print("R2 Score ", r2)
```

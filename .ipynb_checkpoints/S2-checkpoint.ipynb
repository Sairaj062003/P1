```python
from sklearn.preprocessing import LabelEncoder
```

```python
label= LabelEncoder()
```

```python
df1.loc[:, 'transmission'] = label.fit_transform(df1['transmission'])

df1
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
One = OneHotEncoder(sparse_output =False)
```

```python
encoded_fuel = One.fit_transform(df1[['fuel']])
encoded_df = pd.DataFrame(encoded_fuel,columns=One.get_feature_names_out(['fuel']))
df1 = pd.concat([df1,encoded_df],axis=1)
df1
df1=df1.drop(columns = ['fuel'])
df1
```

```python
encoded_seller = One.fit_transform(df1[['seller_type']])
enc_df = pd.DataFrame(encoded_seller,columns=One.get_feature_names_out(['seller_type']))
df1 = pd.concat([enc_df,df1],axis=1)
df1
df1=df1.drop(columns=['seller_type'])
df1
```

```python
from sklearn.model_selection import train_test_split
x=df1['year']
y=df1['km_driven']
print("Initial x shape:", x.shape)
print("Initial y shape:", y.shape)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=40)
```

```python
x_train
x_train.shape
```

```python
y_train
y_train.shape
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python

x_train = x_train.values.reshape(-1, 1)
x_test = x_test.values.reshape(-1, 1)
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
MAE=print("MAE",mean_absolute_error(y_test,y_pred))
```

```python
from sklearn.metrics import mean_squared_error
MAE=print("MSE",mean_squared_error(y_test,y_pred))
```

```python
plt.scatter(df1['year'],df1['selling_price'])
plt.plot(y_test,y_pred,color='red')
plt.xlabel('year')
plt.ylabel('selling_price')
```

```python
from sklearn.metrics import r2_score
print("R2 Score ",r2_score(y_test,y_pred))
```
